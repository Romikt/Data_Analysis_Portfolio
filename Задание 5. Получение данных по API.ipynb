```python
'''
Задание 5. Получение данных по API
Изучите API Геокодера Яндекса
tech.yandex.ru/maps/geocoder/doc/desc/concepts/input_params-docpage/
и получите ключ API для него в кабинете разработчика.
Выполните запрос к API и узнайте долготу точки на карте (Point) для города Самара.
В качестве запасного варианта можно использовать этот ключ - 3f355b88-81e9-4bbf-a0a4-eb687fdea256 - 
он только для выполнения этого задания!
Какая долгота у точки на карте для Самары?
'''
# См. примечания после кода

import requests # Импортирование библиотеки Requests
import json # Импортирование библиотеки Json
r = requests.get("https://geocode-maps.yandex.ru/1.x/?geocode=Самара&apikey=3f355b88-81e9-4bbf-a0a4-eb687fdea256&format=json&result=1") 
# Отправка get-запроса к API
geo = json.loads(r.content) # Получение ответа в формате JSON
print (geo) # Отображение ответа в формате JSON
print ("----------------------------------------------------------------------------------------------------------------")
# Вывод разделяющей строки для лучшего восприятия данных
print("Долгота Самары:", geo['response']['GeoObjectCollection']['featureMember'][0]['GeoObject']['Point']['pos'].split(" ")[0]) 
# Последовательно производится сокращание полученной информации.
# Отображение долготы точки на карте (Point) для города Самара.
```

Необходимо перейти на страницу Геокодера https://yandex.ru/dev/geocode/doc/ru/concepts/input_params и получить ключ API при наличии учётной записи «Яндекса».
Формат запроса: в квадратных скобках - необязательные параметры, остальные обязательные параметры.
apkey – ключ API
•	В «geocode» передаём адрес: Самара
•	format берём Json
•	results - результат ограничим одним объектом = 1
Находим в ответе по поиску «Point» для Самары.
Последовательно сокращаем полученную информацию.
