```python
'''
Задание 6. Получение котировок акций
Получите данные по котировкам акций со страницы:
mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019
и найдите, по какому тикеру был максимальный рост числа сделок (в процентах) за 1 ноября 2019 года.
'''

import requests # Импортирование библиотеки Requests
import pandas as pd # Импортирование библиотеки Pandas
from bs4 import BeautifulSoup # Импортирование библиотеки BeautifulSoup
r = requests.get("https://mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019") 
# Получение данных веб-страницы с помощью get-запроса
html = BeautifulSoup(r.content) # Получение html-кода страницы
print (html) # Отображение html-кода
print ("----------------------------------------------------------------------------------------------------------------")
# Вывод разделяющей строки для лучшего восприятия данных
table = html.find('table', {'id':'marketDataList'}) 
# После просмотра информации какая таблица необходима по тегу table находим id 
# (таргетироваться по нему лучше, но не всегда может сработать)  
rows = [] # Создание массива данных 
trs = table.find_all('tr') # Поиск всех тегов tr внутри таблицы и 
for tr in trs: # перебирая теги
    tr = [td.get_text(strip=True) for td in tr.find_all('td')] # в строку генератором добавляем получение текса из ячеек td, а сами td находятся find _all как теги на строку, удаление ненужных пробелов (strip)    
    if len(tr) > 0: # если размер списка больше 0, 
        rows.append(tr) # к строкам добавляем разобранную строку из таблицы 
print (rows) # Отображение строк
print ("----------------------------------------------------------------------------------------------------------------")
# Вывод разделяющей строки для лучшего восприятия данных
data = pd.DataFrame(rows, columns=["Тикер", "Дата", "Сделки", "C/рост", "С/%", "Закрытие", "Открытие", "min", "max", "avg", "шт", "руб", "Всего"]) 
# загрузка в дата фрейм всего списка списков. Именование столбцов
data = data[data["Сделки"] != "N/A"] # Убираем отсутствующие значения по колонке
data["С/%"] = data["С/%"].str.replace("−","-").str.replace("%","").astype(float) 
# Преобразование столбца %, приведение к числу для сортировки. Удаление (замена) доп. символов
data = data.set_index("С/%") # # Назначение нового индекса по заданному столбцу 
data = data.sort_index(ascending=False) # Сортировка по индексу в порядке убывания
print ("Максимальный рост числа сделок (в процентах) за 1 ноября 2019 года:", data["Тикер"].head(1)) # Вывод максимального значения
```
