```python
'''
Задание 4. Предсказание на 2020 год
Возьмите данные по безработице в городе Москва:
video.ittensive.com/python-advanced/data-9753-2019-07-25.utf.csv
Сгруппируйте данные по годам, и, если в году меньше 6 значений, отбросьте эти годы.
Постройте модель линейной регрессии по годам среднего значения отношения UnemployedDisabled к UnemployedTotal 
(процента людей с ограниченными возможностями) 
за месяц и ответьте, какое ожидается значение процента безработных инвалидов в 2020 году 
при сохранении текущей политики города Москвы?
Ответ округлите до сотых. Например, 2,32
Какое ожидается значение в 2020 году при сохранении текущей политики города Москвы?
'''

import pandas as pd # Импортирование библиотеки Pandas
import numpy as np # Импортирование библиотеки Numpy
from sklearn.linear_model import LinearRegression # Импортирование из библиотеки Sklearn модуля LinearRegression
data = pd.read_csv("https://video.ittensive.com/python-advanced/data-9753-2019-07-25.utf.csv", delimiter=";") 
# Загрузка csv-файла и разделение столбцов по ";" в строках CSV-файла
print (data) # Вывод данных Pandas из переменной data
print ("----------------------------------------------------------------------------------------------------------------")
# Вывод разделяющей строки для лучшего восприятия данных
data["UDP"] = 100*data["UnemployedDisabled"]/data["UnemployedTotal"] # Создание отдельного столбца после вычислений
# Вычисление в % кол-ва людей с ограниченными возможностями от общего числа безработных для построения модели лин. рег.
data_group = data.groupby("Year").filter(lambda x: x["UDP"].count() > 5)  
# Группировка данных по столбцу "Year" и фильтрация этого столбца с применением лямбда ф-ции, где подсчёт значений больше 5
data_group = data_group[["UDP", "Year"]].groupby("Year").mean()
# Группировка значений по столбцу "Year" и вычисление среднего арифметического
x = np.array(data_group.index).reshape(len(data_group.index), 1) 
# Создание массива данных по сформировнному и привидение индексов групп данных к двумерному массиву
y = np.array(data_group["UDP"]).reshape(len(data_group.index), 1) 
# Создание массива данных по данным столбца "UDP" и привидение индексов групп данных к двумерному массиву
model = LinearRegression() # Построение модели линейной регрессии
model.fit(x, y) # Вычисление коэффициентов линейной регрессии
print ("Ожидаемое значение в 2020 году процента безработных инвалидов при сохранении текущей политики города Москвы:", np.round(model.predict(np.array(2020).reshape(1, 1)), 2)) 
# Отображение предсказанного значения на 2020 год
```
