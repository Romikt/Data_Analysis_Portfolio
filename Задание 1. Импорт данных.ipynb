```python
'''
Задание 1. Импорт данных
Возьмите данные по вызовам пожарных служб в Москве за 2015-2019 годы:
https://video.ittensive.com/python-advanced/data-5283-2019-10-04.utf.csv
Получите из них фрейм данных (таблицу значений). По этому фрейму вычислите среднее значение вызовов пожарных машин в месяц 
в одном округе Москвы, округлив до целых
Примечание: найдите среднее значение вызовов, без учёта года
'''

import pandas as pd # Импортирование библиотеки Pandas
data = pd.read_csv("https://video.ittensive.com/python-advanced/data-5283-2019-10-04.utf.csv", delimiter=";") 
# Загрузка csv-файла и разделение столбцов по ";" в строках CSV-файла
print (data.head) # Просмотр первых и последних строк кадра данных Pandas 
print () # Пропуск строки
print ("Округлённое до целых среднее значение по вызовам пожарных служб в Москве за 2015-2019 г. без учёта года: ", data["Calls"].mean().round())
```
